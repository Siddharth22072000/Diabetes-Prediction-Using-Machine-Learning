```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, r2_score, classification_report, roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
le = LabelEncoder()

df['gender'] = le.fit_transform(df['gender'])
df['smoking_history'] = le.fit_transform(df['smoking_history'])

df.head()
```

```python
X = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=27, test_size=0.3)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(X_train, y_train)
```

```python
knn_pred = knn.predict(X_test)
```

```python
print(r2_score(y_test, knn_pred))
print('\n')
print(confusion_matrix(y_test, knn_pred))
print('\n')
print(classification_report(y_test, knn_pred))
```

```python
knn_pred_proba = knn.predict_proba(X_test)

print(roc_auc_score(y_test, knn_pred_proba[:,1]))
```

```python
#RAMDOM FOREST
rcf= RandomForestClassifier(n_estimators=1600)
```

```python
rcf.fit(X_train, y_train)
```

```python
rcf_pred = rcf.predict(X_test)
```

```python
print(r2_score(y_test, rcf_pred))
print('\n')
print(confusion_matrix(y_test, rcf_pred))
print('\n')
print(classification_report(y_test, rcf_pred))
```

```python
rcf_pred_proba = rcf.predict_proba(X_test)

print(roc_auc_score(y_test, rcf_pred_proba[:,1]))
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
tr_pred = dtree.predict(X_test)
```

```python
print(r2_score(y_test, tr_pred))
print('\n')
print(confusion_matrix(y_test, tr_pred))
print('\n')
print(classification_report(y_test, tr_pred))
```

```python
tr_pred_proba = dtree.predict_proba(X_test)

print(roc_auc_score(y_test, tr_pred_proba[:,1]))
```

```python
###Support Vector Classifier
svc = SVC(kernel='linear', probability=True)
```

```python
# Train the SVC on the training data
svc.fit(X_train, y_train)
```

```python
# Predicting the labels of the test data
svc_pred = svc.predict(X_test)
```

```python
print(r2_score(y_test, svc_pred))
print('\n')
print(confusion_matrix(y_test, svc_pred))
print('\n')
print(classification_report(y_test, svc_pred))
```

```python
# Printing the accuracy score of the classifier
print("Accuracy:", svc.score(X_test, y_test))
```

```python
## Combing the results
```

```python
from sklearn.ensemble import VotingClassifier

# Create instances of the individual classifiers
knn = KNeighborsClassifier(n_neighbors=7)
rcf = RandomForestClassifier(n_estimators=1600)
dtree = DecisionTreeClassifier()

# Create a voting classifier that combines the predictions of the three classifiers
voting_classifier = VotingClassifier(estimators=[
    ('knn', knn),
    ('rcf', rcf),
    ('dtree', dtree)
], voting='hard')  # 'hard' means majority class voting

# Fit the voting classifier on the training data
voting_classifier.fit(X_train, y_train)

# Make predictions using the voting classifier
voting_pred = voting_classifier.predict(X_test)

# Evaluate the performance of the voting classifier

print('\n')
print("Voting Classifier R2 Score:", r2_score(y_test, voting_pred))
print('\n')
print("Confusion Matrix for Voting Classifier:\n", confusion_matrix(y_test, voting_pred))
print('\n')
print("Classification Report for Voting Classifier:\n", classification_report(y_test, voting_pred))

```

```python
### Importing SMOTE
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=5)
```

```python
## Creating balanced data
```

```python
x_bal, y_bal = sm.fit_resample(X, y)
```

```python
## Splitting balanced data into Training and Testing data
```

```python
X_bal_train, X_bal_test, y_bal_train, y_bal_test = train_test_split(x_bal, y_bal, test_size=0.3, random_state=42)
```

```python
##KNN Model on Balanced Data
```

```python
knn1 = KNeighborsClassifier(2)
knn1.fit(X_bal_train, y_bal_train)

knn1_pred = knn1.predict(X_bal_test)
```

```python
print(r2_score(y_bal_test, knn1_pred))
print('\n')
print(confusion_matrix(y_bal_test, knn1_pred))
print('\n')
print(classification_report(y_bal_test, knn1_pred))
```

```python

knn1_pred_proba = knn1.predict_proba(X_bal_test)

print(roc_auc_score(y_bal_test, knn1_pred_proba[:,1]))
```

```python
## Random Forest Model on Balanced Data
```

```python
rcf_b = RandomForestClassifier(n_estimators=100)

rcf_b.fit(X_bal_train, y_bal_train)

rcf_b_pred = rcf_b.predict(X_bal_test)
```

```python
print(r2_score(y_bal_test, rcf_b_pred))
print('\n')
print(confusion_matrix(y_bal_test, rcf_b_pred))
print('\n')
print(classification_report(y_bal_test, rcf_b_pred))
```

```python
rcf_b_pred_proba = rcf_b.predict_proba(X_bal_test)

print(roc_auc_score(y_bal_test, rcf_b_pred_proba[:,1]))
```

```python
## Decision Tree Model on Balanced Data
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree_b = DecisionTreeClassifier()
dtree_b.fit(X_bal_train, y_bal_train)
dtree_b_pred = dtree_b.predict(X_bal_test)
```

```python
print(r2_score(y_bal_test, dtree_b_pred))
print('\n')
print(confusion_matrix(y_bal_test, dtree_b_pred))
print('\n')
print(classification_report(y_bal_test, dtree_b_pred))
```

```python
dtree_b_pred_proba = dtree_b.predict_proba(X_bal_test)

print(roc_auc_score(y_bal_test, dtree_b_pred_proba[:,1]))
```

```python
#Combing the above result
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score

```

```python

knn_model = knn1
rf_model = rcf_b
dt_model = dtree_b


```

```python
# Create a Voting Classifier
voting_classifier = VotingClassifier(estimators=[
    ('knn', knn_model),
    ('rf', rf_model),
    ('dt', dt_model)
], voting='hard')

```

```python
# Fit the Voting Classifier on balanced training data
voting_classifier.fit(X_bal_train, y_bal_train)

```

```python
# Make predictions on balanced test data
ensemble_predictions = voting_classifier.predict(X_bal_test)

```

```python
# Calculate the accuracy of the ensemble predictions
ensemble_accuracy = accuracy_score(y_bal_test, ensemble_predictions)

```

```python
print("Ensemble Accuracy:", ensemble_accuracy)
```
